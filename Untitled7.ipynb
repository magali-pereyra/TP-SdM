```python
class Persona():
    def __init__(self, nombre, apellido, cedula):
        if not isinstance(nombre, str):
            raise ValueError("El nombre no es válido")
        if not isinstance(apellido, str):
            raise ValueError("El apellido no es válido")

        try:
            cedula= int(cedula)
        except ValueError:
            raise ValueError("La cédula no es válida")

        self.nombre = nombre
        self.apellido = apellido
        self._cedula = cedula

    @property
    def cedula(self):
        return self._cedula

    def nombre_completo(self):
        return f"{self.nombre} {self.apellido}"

persona=Persona("Julia","Iturriaga",12)
```

```python
import json
class UnidadCurricular:
    def __init__(self, codigo, nombre, creditos, previas):
        self.codigo = codigo
        self.nombre = nombre
        self.creditos = creditos
        self.previas = previas

class PlanDeEstudio:
    def __init__(self, nombre_plan):
        self.nombre_plan = nombre_plan
        self.materias = []

    def agregar_materia(self, unidad_curricular):
        self.materias.append(unidad_curricular)

    def buscar_uc_por_codigo(self, codigo):
        for uc in self.materias:
            if uc.codigo == codigo:
                return uc
        return None

    @classmethod
    def desde_json(cls, ruta_archivo):
        with open(ruta_archivo, "r", encoding="utf-8") as f:
            datos = json.load(f)

        plan = cls("Plan de Estudio")
        for semestre, materias in datos.items():
            for codigo, materia in materias.items():
                uc = UnidadCurricular(
                    codigo=codigo,
                    nombre=materia["nombre"],
                    creditos=materia["creditos"],
                    previas=materia["previas"]
                )
                plan.agregar_materia(uc)
        return plan
plan=PlanDeEstudio.desde_json("archivo UCS.json")
for materia in plan.materias:
    print(f"{materia.codigo} - {materia.nombre} ({materia.creditos} créditos)")
    if materia.previas:
        print(materia.previas)
```

```python
import json

class Estudiante(Persona):
    def __init__(self, nombre, apellido, cedula, año_ingreso,plan):
        super().__init__(nombre, apellido, cedula)
        self.año_ingreso = año_ingreso
        self.plan=plan
        self.ucs_aprobadas = []
        self.ucs_regulares = []
        self.ucs_a_examen = []
        self.ucs_cursando = []

    def ver_ucs_aprobadas(self):
        if not self.ucs_aprobadas:
            print("No tenés UCs aprobadas.")
        else:
            print("UCs aprobadas:")
            for uc in self.ucs_aprobadas:
                print(f"- {uc.codigo}: {uc.nombre}")

    def ver_ucs_regulares(self):
        if not self.ucs_regulares:
            print("No estás regular en ninguna UC.")
        else:
            print("UCs regulares:")
            for uc in self.ucs_regulares:
                print(f"- {uc.codigo}: {uc.nombre}")

    def ver_ucs_cursando(self):
        if not self.ucs_cursando:
            print("No estás cursando ninguna UC.")
        else:
            print("UCs que estás cursando:")
            for uc in self.ucs_cursando:
                print(f"- {uc.codigo}: {uc.nombre}")

    def puede_inscribirse(self, uc):
        codigos_aprobadas = [u.codigo for u in self.ucs_aprobadas]
        codigos_cursando = [u.codigo for u in self.ucs_cursando]
        previas_faltantes = []

        for previa_codigo in uc.previas:
          previa = self.plan.buscar_uc_por_codigo(previa_codigo)
          if previa.codigo not in codigos_aprobadas:
            previas_faltantes.append(previa)

        if previas_faltantes:
          print(f"No puedes inscribirte en la UC {uc.nombre} porque te faltan las siguientes previas:")
        for previa in previas_faltantes:
            print(f"- {previa.nombre}")
        return False

        if uc.codigo in codigos_cursando:
            print("Ya está cursando esta UC.")
            return False

        if uc.codigo in codigos_aprobadas:
            print("Ya aprobó esta UC.")
            return False
        return True

    def matricularse_a_uc(self, codigo_uc):
        uc = self.plan.buscar_uc_por_codigo(codigo_uc)
        if not uc:
            print("No se encontró la UC.")
            return

        if self.puede_inscribirse(uc):
            self.ucs_cursando.append(uc)
            with open("uc_inscriptos.txt", "a", encoding="utf-8") as f:
                f.write(f"{self.nombre} {self.apellido} - Cédula: {self.cedula} - UC: {uc.codigo} {uc.nombre}\n")
            print(f"Te inscribiste a {uc.nombre}")

    def matricularse_a_examen(self, codigo_uc):
        uc = self.plan.buscar_uc_por_codigo(codigo_uc)
        if not uc:
            print("No se encontró la UC.")
            return

        if uc in self.ucs_regulares:
            self.ucs_a_examen.append(uc)
            with open("examen_inscriptos.txt", "a", encoding="utf-8") as f:
                f.write(f"{self.nombre} {self.apellido} - Cédula: {self.cedula} - UC: {uc.codigo} {uc.nombre}\n")
            print(f"Te inscribiste al examen de {uc.nombre}")
        else:
            print(f"No podés inscribirte al examen de {uc.nombre}: no estás regular.")

estudiante=Estudiante("Julia","Iturriaga",12,2022,plan)
estudiante.matricularse_a_uc("UC2S5")

```

```python
class Secretaria(Persona):
    def __init__(self, nombre, apellido, cedula):
        super().__init__(nombre, apellido, cedula)
        self.ucs_aprobadas = []
        self.ucs_regulares = []
        self.ucs_a_examen = []
        self.ucs_cursando = []

    def puede_inscribirse(self, uc, plan):
        codigos_aprobadas = [u.codigo for u in self.ucs_aprobadas]
        codigos_cursando = [u.codigo for u in self.ucs_cursando]

        for previa in uc.previas:
            if previa not in codigos_aprobadas:
                print(f"No cumple con la previatura: {previa}")
                return False

        if uc.codigo in codigos_cursando:
            print("Ya está cursando esta UC.")
            return False
        if uc.codigo in codigos_aprobadas:
            print("Ya aprobó esta UC.")
            return False

        return True

    def matricularse_a_uc(self, uc, plan):
        archivo_uc = "uc_inscriptos.txt"  # Archivo para estudiantes matriculados a la UC
        if self.puede_inscribirse(uc, plan):
            self.ucs_cursando.append(uc)
            print(f"Matriculado/a exitosamente en {uc.nombre}.")
            with open(archivo_uc, "a", encoding="utf-8") as f:
                f.write(f"{self.nombre} {self.apellido} - Cédula: {self.cedula} - UC: {uc.codigo} {uc.nombre}\n")
        else:
            print("No puede inscribirse a esta UC.")

    def matricularse_a_examen(self, uc, plan):
        archivo_examen = "examen_inscriptos.txt"
        if uc in self.ucs_regulares:
            print("Se ha matriculado al examen con éxito.")
            self.ucs_a_examen.append(uc)
            with open(archivo_examen, "a", encoding="utf-8") as f:
                f.write(f"{self.nombre} {self.apellido} - Cédula: {self.cedula} - UC: {uc.codigo} {uc.nombre}\n")
        else:
            print("No puede inscribirse al examen porque no es regular en esa UC.")

    def desmatricular_examen(self,uc,plan):
        archivo_examen= "examen_inscriptos.txt"
        if uc in self.ucs_a_examen:
            self.ucs_a_examen.remove(uc)
            print(f"Desmatriculado/a del examen de {uc.nombre} con éxito.")
            with open(archivo_examen, "r", encoding="utf-8") as f:
                lineas = f.readlines()
            with open(archivo_examen, "w", encoding="utf-8") as f:
                for linea in lineas:
                    if f"{self.nombre} {self.apellido} - Cédula: {self.cedula} - UC: {uc.codigo} {uc.nombre}" not in linea:
                        f.write(linea)
        else:
            print("No está inscrito/a en este examen.")

    def desmatricular_uc(self, uc):
        archivo_uc = "uc_inscriptos.txt"
        if uc in self.ucs_cursando:
            self.ucs_cursando.remove(uc)
            print(f"Desmatriculado/a de {uc.nombre} con éxito.")
            with open(archivo_uc, "r", encoding="utf-8") as f:
                lineas = f.readlines()
            with open(archivo_uc, "w", encoding="utf-8") as f:
                for linea in lineas:
                    if f"{self.nombre} {self.apellido} - Cédula: {self.cedula} - UC: {uc.codigo} {uc.nombre}" not in linea:
                        f.write(linea)
        else:
            print(f"No está matriculado/a en {uc.nombre}.")

    def agregar_uc_aprobada(self, uc_nueva):
        if uc_nueva not in self.ucs_aprobadas:
            self.ucs_aprobadas.append(uc_nueva)

    def agregar_uc_regular(self, uc_nueva):
        if uc_nueva not in self.ucs_regulares:
            self.ucs_regulares.append(uc_nueva)
```
